```python
import numpy as np
import matplotlib.pyplot as p
font = {'family': 'serif',
        'color':  'orange',
        'weight': 'bold',
        'size': 24,
        }
name = ['1st ', '2nd', '3rd', '4th', '5th','6th', '7th', '8th', '9th', '10th', '11th', '12th']

cgpa = [3.50,3.153,3.278,3.168,3.210,3.295,3.378,3.443,3.510,3.601,3.610,3.578]
gpa = [3.50,2.837,3.481,2.880,3.361,3.650,3.852,3.663,3.520,3.750,3.652,3.456]

p.plot(name, cgpa, color="blue", label='cgpa', marker="*")
p.plot(name, gpa, color="green", label='gpa', marker="+")

p.xlabel('Semester',fontdict=font)
p.ylabel('GPA\CGPA',fontdict=font)
p.title("Visulization Result by Semester",fontdict=font)
p.legend()
p.show()
```
